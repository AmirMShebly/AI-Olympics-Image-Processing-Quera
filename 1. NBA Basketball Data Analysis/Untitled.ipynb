***Get a list of players and seasons in which that player recorded double-digit stats (greater than 10) in at least 2 of the 5 major criteria.
The 5 main criteria are average points, average number of rebounds (sum of rebounds in defense and attack), average number of passes leading to goals, average number of blocks and average number of steals.***

```python
data['REB'] = data['OREB'] + data['DREB']
rules = [
    data['PTS'] >= 10,
    data['REB'] >= 10,
    data['AST'] >= 10,
    data['STL'] >= 10,
    data['BLK'] >= 10,
]
data['double_double_rules'] = sum(rules)
double_double = data[data['double_double_rules']>=2]

double_double = double_double[['PLAYER','YEAR','PTS','AST','REB','STL','BLK']].sort_values(by=['YEAR', 'PLAYER'])
```

***Determine the most valuable players of these 10 seasons. The value of a player in a season is obtained using the following formula:***

***(ùëÉùëáùëÜ + ùëÇùëÖùê∏ùêµ + ùê∑ùëÖùê∏ùêµ + ùê¥ùëÜùëá + ùëÜùëáùêø + ùêµùêøùêæ) ‚àí (ùëáùëÇùëâ + ùëÄùëñùë†ùë†ùëíùëëùêπùê∫ + ùëÄùëñùë†ùë†ùëíùëëùêπùëá)***

***Missed FG and Missed FT mean the average number of missed 2-point and free throws, respectively.
The value of a player in all 10 seasons is equal to his average value in each season.***

```python
data['Missed_FG'] = data['FGA'] - data['FGM']
data['Missed_FT'] = data['FTA'] - data['FTM']

data['VALUE'] = (
    data['PTS'] +
    data['OREB'] +
    data['DREB'] +
    data['AST'] +
    data['STL'] +
    data['BLK'] -
    data['TOV'] -
    data['Missed_FG'] -
    data['Missed_FT']
)

best_player = data.groupby('PLAYER')['VALUE'].mean().reset_index()
best_player['VALUE'] = best_player['VALUE'].round(2)
best_player = best_player.sort_values(by=['VALUE','PLAYER'],ascending=[False,True])
```

```python
best_player.head()
```

***Which team has the highest scoring average in each year?***


***Note that the average score of a team is equal to the total average score of its players***

```python
team_points = data.groupby(['YEAR','TEAM'])['PTS'].sum().reset_index()
max_points = team_points.loc[team_points.groupby('YEAR')['PTS'].idxmax()]
max_points['PTS'] = max_points['PTS'].round(2)
max_PTS_of_year = max_points.sort_values(by='YEAR')
```

```python
import zipfile

if not os.path.exists(os.path.join(os.getcwd(), 'NBA.ipynb')):
    %notebook -e NBA.ipynb

double_double.to_csv('double_double.csv', index=False)
best_player.to_csv('best_player.csv', index=False)
max_PTS_of_year.to_csv('max_PTS_of_year.csv', index=False)

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

file_names = ['double_double.csv', 'best_player.csv', 'max_PTS_of_year.csv', 'NBA.ipynb']
compress(file_names)
```
